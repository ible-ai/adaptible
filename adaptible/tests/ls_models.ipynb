```python
from huggingface_hub import HfApi

api = HfApi()
# List models with a specific filter (e.g., "automatic-speech-recognition")
models = list(api.list_models(filter="text-generation") )
```

```python
import datetime

filtered_models = []
for model in sorted(models, key=lambda x: x.downloads, reverse=True):
    model_name = str(model.modelId).lower()
    # 
    delta = datetime.timedelta(30 * 5, 0, 0) # < 5 months old.
    # if "qwen" in model_name and "mlx" in model_name and "4b" in model_name:
    if "mlx" in model.tags and datetime.datetime.now().astimezone() - model.created_at < delta:
        # if "14b" in model_name or "4bit" in model_name: continue
        # if "14b" in model_name or "4bit" in model_name: continue
        if model.downloads < 1000: continue
        filtered_models.append(model)
        print(model.downloads, "\t", model.modelId)
# Use code with caution.
```
